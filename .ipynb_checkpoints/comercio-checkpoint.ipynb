```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
nyt = 'https://diariocorreo.pe/politica/'
```

```python
resultado = requests.get(nyt)
```

```python
contenido = resultado.text
```

```python
soup = BeautifulSoup(contenido, 'html.parser')
```

```python
print(soup.prettify())
```

```python
noticia = soup.find_all('div', class_='stories-news w-full')
print(noticia)
```

```python
noticias = list()
for i in noticia:
    noticias.append(i.text)
    
print(noticias)
```

```python
encabezado =soup.find_all('h2', class_='story-item__content-title overflow-hidden')
print(encabezado)
```

```python
encabezados = list()
for i in encabezado:
    encabezados.append(i.text)
    
print(encabezados)
```

```python
resumen = soup.find_all('p', class_='story-item__subtitle overflow-hidden hidden mt-10 mb-10 text-md text-gray-200 line-h-xs')
print(resumen)
```

```python
resumenes = list()
for i in resumen:
    resumenes.append(i.text)
    
print(resumenes)
```

```python
data = pd.DataFrame({'Noticia':encabezados, 'Cuerpo':resumenes,})
data
```
