```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
nyt = 'https://elcomercio.pe/ultimas-noticias/'
```

```python
resultado = requests.get(nyt)
```

```python
contenido = resultado.text
```

```python
soup = BeautifulSoup(contenido, 'html.parser')
```

```python
print(soup.prettify())
```

```python
noticia = soup.find_all('div', class_='story-item__information-box w-full')
print(noticia)
```

```python
noticias = list()
for i in noticia:
    noticias.append(i.text)
    
print(noticias)
```

```python
encabezado =soup.find_all('h2', class_='story-item__content-title overflow-hidden')
print(encabezado)
```

```python
encabezados = list()
for i in encabezado:
    encabezados.append(i.text)
    
print(encabezados)
```

```python
fecha = soup.find_all('p', class_='story-item__date font-thin ml-5 text-xs text-gray-300 md:mt-5 md:ml-0')
print()
```

```python
fechas = list()
for i in fecha:
    fechas.append(i.text)
    
print(fechas)
```

```python
autor = soup.find_all('div', class_='hidden')
```

```python
autores = list()
for i in autor:
    autores.append(i.text)
print(autores)
```

```python
data = pd.DataFrame({'Noticia':encabezados, 'Fecha y Hora':fechas, 'Autor': autores})
data
```

```python
data.info()
```

```python
pts_club=data.loc[:,['Noticia','Autor']]
pts_club=pts_club.sort_values('Autor', ascending=False)
pts_club
```
